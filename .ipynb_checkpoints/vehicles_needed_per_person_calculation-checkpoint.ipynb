```python
!pip install ortools
from ortools.linear_solver import pywraplp
import json
import requests

response = requests.get("https://gitlab.com/drvicsana/cop-proyecto-2023/-/raw/main/project_data/incidentes2019.json")
incidents_db = json.loads(response.text)

response = requests.get("https://gitlab.com/drvicsana/cop-proyecto-2023/-/raw/main/project_data/distancias_estaciones_barrios.json")
distances_stations_ntas_db = json.loads(response.text)

response = requests.get("https://gitlab.com/drvicsana/cop-proyecto-2023/-/raw/main/project_data/distancias_estaciones.json")
distances_stations_db = json.loads(response.text)

response = requests.get("https://gitlab.com/drvicsana/cop-proyecto-2023/-/raw/main/project_data/barrios.json")
ntas_db = json.loads(response.text)

response = requests.get("https://gitlab.com/drvicsana/cop-proyecto-2023/-/raw/main/project_data/estaciones.json")
stations_db = json.loads(response.text)
```

```python
# iterating all neighborhoods
for neighbor in ntas_db:
    
    vehicle_types = ["engine", "ladder", "rescue", "squad", "hazardous"]
    
    neighbor["vehicles_needed_per_person_first_shift"] = {}
    neighbor["vehicles_needed_per_person_second_shift"] = {}
    
    population = neighbor["population"]
    vehicles_needed_per_person_first_shift = neighbor["vehicles_needed_per_person_first_shift"]
    vehicles_needed_per_person_second_shift = neighbor["vehicles_needed_per_person_second_shift"]
    
    for vehicle_type in vehicle_types:
        vehicles_needed_per_person_first_shift[vehicle_type] = 0
        vehicles_needed_per_person_second_shift[vehicle_type] = 0

    for incident in incidents_db:
        if incident["nta"] == neighbor["nta"]:
            vehicles_needed_for_incident = incident["units"]
            is_first_shift = incident["is_first_shift"]
            
            if is_first_shift == True:
                for vehicle_type in vehicle_types:
                    vehicles_needed_per_person_first_shift[vehicle_type] += vehicles_needed_for_incident.count(vehicle_type)
                    
            else:
                for vehicle_type in vehicle_types:
                    vehicles_needed_per_person_second_shift[vehicle_type] += vehicles_needed_for_incident.count(vehicle_type)
    
    if population != 0:
        for vehicle_type in vehicle_types:
            vehicles_needed_per_person_first_shift[vehicle_type] /= population
            vehicles_needed_per_person_second_shift[vehicle_type] /= population
```

```python
for neighbor in ntas_db:
    # calculate population density
    population = neighbor["population"]
    
    shape_area = neighbor["shape_area"] / 10000000
    
    neighbor["population_density"] = population / shape_area
```

```python
for neighbor in ntas_db:
    
    vehicle_types = ["engine", "ladder", "rescue", "squad", "hazardous"]
    
    population_density = neighbor["population_density"]
    
    vehicles_needed_per_person_first_shift = neighbor["vehicles_needed_per_person_first_shift"]
    vehicles_needed_per_person_second_shift = neighbor["vehicles_needed_per_person_second_shift"]
    
    neighbor["vehicles_needed_per_density_first_shift"] = {}
    neighbor["vehicles_needed_per_density_second_shift"] = {}
    
    vehicles_needed_per_density_first_shift = neighbor["vehicles_needed_per_density_first_shift"]
    vehicles_needed_per_density_second_shift = neighbor["vehicles_needed_per_density_second_shift"]
    
    for vehicle_type in vehicle_types:
        vehicles_needed_per_density_first_shift[vehicle_type] = vehicles_needed_per_person_first_shift[vehicle_type] * population_density
        vehicles_needed_per_density_second_shift[vehicle_type] = vehicles_needed_per_person_second_shift[vehicle_type] * population_density
```

```python
ranks_stations_ntas_db = distances_stations_ntas_db
ntas = [neighborhood["nta"] for neigborhood in ntas_db]

for station in ranks_stations_ntas_db:
    for nta in ntas:
        for neigborhood in ntas_db:
            if neigborhood["nta"] == nta:
                station[nta] = neigborhood
```
