```python
#df_ntas = pd.read_csv("dataset/neighborhoods_info_standardized.csv", index_col="nta")
df_distance_stations_ntas = pd.read_csv("dataset/distance_stations_neighborhoods.csv", index_col="Unnamed: 0")
df_distance_stations = pd.DataFrame(distances_stations_db)
df_distance_stations
```

```python
df_neighborhoods_serviceability = df_distance_stations_ntas.copy()
df_neighborhoods_serviceability.loc[:] = False
df_neighborhoods_serviceability
```

```python
# interate together the neigborhoods on nta-stations distances dataframe and neighborhoods dataframe
for (nta, distances) in df_distance_stations_ntas.iterrows():
    # iterate stations' distances from the neighborhood
    for station, distance_st_nta in distances.items():
        # check if the distance between the nta and the station is less then the 20 percentile
        if distance_st_nta <= 754:
            # set the neighborhood as servable by that station
            df_neighborhoods_serviceability.loc[nta, station] = True
    
    # check that a neighborhood is served at least by one station (in case we have to change the percentile)
    assert any(df_neighborhoods_serviceability.loc[nta])
```
